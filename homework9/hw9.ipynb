```python
# Your code here
import matplotlib.pyplot as plt
import numpy as np 

# Defining x and y values plus arrays for the dot plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


def graph(x, y):
    # This function will take the x and y values and graph them in both a line graph and dot plot

    fig, ax = plt.subplots(1, 2, figsize=(10, 10))
    # Line graph
    ax[0].plot(x, y, label="f(x) = 2x", linestyle='-', color="red")
    plt.title("X vs. Y")
    plt.xlabel("X-Axis", fontsize=15)
    plt.ylabel("Y-Axis", fontsize=15)
    # Dot graph
    ax[1].scatter(x, y, marker='o', color="blue")
    plt.title("X vs. Y")
    plt.xlabel("X-Axis", fontsize=15)
    plt.ylabel("Y-Axis", fontsize=15)
    
    # Adding legend and showing the graph
   
    plt.legend()
    plt.show()
    return

graph(x,y)
```

# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title

- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
# Your code here
# Defining x values
x_data = np.linspace(0, 10, 100)

# Defining functions
f = np.power(x_data, 2)
g = np.sqrt(x_data)

def graph(x_data):
    # This function will take the x and both output functions and graph them in in the same plot
    figsize=(10,8)
    plt.title("Plot of f(x) and g(x)")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.plot(x_data, f, label="f(x)=x^2", linestyle='-', color="blue")
    plt.plot(x_data, g, label="g(x)=sqrt(x)", linestyle='--', color="red")
    
    # Adding legend and showing the graph
    plt.legend()
    plt.show()
    return

graph(x_data)
```

```python
# Import your module and/or your code here
import numpy as np
import matplotlib.pyplot as plt

# Defining the range of theta and functions for the three axes
theta = np.linspace(0, 2*np.pi)
x = np.cos(theta - np.pi/2) * np.sin(theta)
y = np.sin(theta - np.pi/2)
z = theta

def stem_plot_3d(x, y, z):
    # Projecting the 3D stemplot
    fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
    
    # Customizing the appearance of the lines and points and the position of the baseline
    markerline, stemlines, baseline = ax.stem(x, y, z, linefmt='blue', markerfmt='D', bottom=np.pi)
    markerline.set_markerfacecolor('none')

    # Title and labels
    ax.set(title="3D Stem Plot")
    ax.set(xlabel="X-axis", ylabel="Y-label", zlabel="z_axis")

    plt.show()
    return

stem_plot_3d(x, y, z)
```
